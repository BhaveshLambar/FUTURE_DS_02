```python
date_cols_accounts = ['signup_date']
date_cols_subs = ['start_date', 'end_date']
date_cols_churn = ['churn_date']

for col in date_cols_accounts: 
    accounts[col] = pd.to_datetime(accounts[col])

for col in date_cols_subs:
    subscriptions[col] = pd.to_datetime(subscriptions[col])

for col in date_cols_churn:
    churn_events[col] = pd.to_datetime(churn_events[col])
    
```

```python
master = accounts.copy()
master.shape
```

```python
subs_agg = (
    subscriptions
    .groupby('account_id')
    .agg(
        subscription_start = ('start_date', 'min'),
        subscription_end = ('end_date', 'max'),
        total_mrr = ('mrr_amount', 'sum'),
        total_arr = ('arr_amount', 'sum'),
        has_upgraded = ('upgrade_flag', 'max'),
        has_downgraded = ('downgrade_flag', 'max'),
        billing_frequency = ('billing_frequency', lambda x: x.mode().iloc[0]),
        current_plan = ('plan_tier', lambda x: x.iloc[-1])
    )
    .reset_index()
)

subs_agg.head(5)
```

```python
churn_agg = (
    churn_events
    .groupby('account_id')
    .agg(
        churned = ('account_id', 'count'),
        churned_date = ('churn_date', 'max'),
        churned_reason = ('reason_code', lambda x: x.mode().iloc[0]),
        refund_amount = ('refund_amount_usd', 'sum'),
        reactivated = ('is_reactivation', 'max')
    )
    .reset_index()
)

churn_agg['churned'] = churn_agg['churned'] > 0
churn_agg.head(5)
```

```python
usage_agg = (
    feature_usage
    .groupby('subscription_id')
    .agg(
        total_usage = ('usage_count', 'sum'),
        total_time = ('usage_duration_secs', 'sum'),
        total_errors = ('error_count', 'sum'),
        total_beta = ('is_beta_feature', 'max')
    )
    .reset_index()
)

usage_agg.head(5)
```

```python
usage_agg = usage_agg.merge(
    subscriptions[['subscription_id', 'account_id']],
    on = 'subscription_id',
    how = 'left'
)
```

```python
usage_agg = (
    usage_agg
    .groupby('account_id')
    .agg(
        total_usage = ('total_usage', 'sum'),
        total_time = ('total_time', 'sum'),
        total_errors = ('total_errors', 'sum'),
        used_beta = ('total_beta', 'max')
    )
    .reset_index()
)

usage_agg.head(5)
```

```python
tickets_agg = (
    support_tickets
    .groupby('account_id')
    .agg(
        total_tickets = ('ticket_id', 'count'),
        avg_resolution_time = ('resolution_time_hours', 'mean'),
        high_priority_tickets = ('priority', lambda x: (x.isin(['high', 'urgent'])).sum()),
        escalations = ('escalation_flag', 'sum'),
        avg_satisfaction = ('satisfaction_score', 'mean')
    )
    .reset_index()
)
tickets_agg.head(5)
```

```python
master = master.merge(subs_agg, on = 'account_id', how = 'left')
master = master.merge(churn_agg, on = 'account_id', how = 'left')
master = master.merge(usage_agg, on = 'account_id', how = 'left')
master = master.merge(tickets_agg, on = 'account_id', how = 'left')

master.shape
```

```python
master.info()
```

```python
master['churned'] = master['churned'].fillna(False)
master['reactivated'] = master['reactivated'].fillna(False)

numeric_cols = [
    'total_mrr', 'total_arr', 'total_usage', 'total_time', 'total_tickets', 'total_errors', 'escalations'
]
master[numeric_cols] = master[numeric_cols].fillna(0)
```

```python
analysis_date = pd.to_datetime('today')

master['end_date_final'] = master['churned_date'].fillna(analysis_date)

master['tenure_days'] = (
    master['end_date_final'] - master['signup_date']
).dt.days
```

### Overall Churn Rate

```python
total_customers = master.shape[0]
churned_customers = master['churned'].sum()

churn_rate = churned_customers / total_customers
churn_rate
```

### Churn Rate by Plan Tier

```python
churn_by_plan = (
    master
    .groupby('current_plan')
    .agg(
        customers = ('account_id', 'count'),
        churned = ('churned', 'sum')
    )
    .reset_index()
)

churn_by_plan['churn_rate'] = churn_by_plan['churned'] / churn_by_plan['customers']
churn_by_plan
```

```python
master['cohort_month'] = master['signup_date'].dt.to_period('M').dt.to_timestamp()
```

```python
master['tenure_months'] = (master['tenure_days'] / 30).astype(int)
```

```python
cohort_counts = (
    master
    .groupby(['cohort_month', 'tenure_months'])
    .agg(customers = ('account_id', 'nunique'))
    .reset_index()
)

cohort_counts.head()
```

```python
cohort_pivot = cohort_counts.pivot(
    index="cohort_month",
    columns="tenure_months",
    values="customers"
)
```

```python
cohort_size = cohort_pivot.iloc[:, 0]

retention_matrix = cohort_pivot.divide(cohort_size, axis=0)
```

```python
retention_matrix = retention_matrix.loc[:, :12]
```

```python
master["tenure_bucket"] = pd.cut(
    master["tenure_months"],
    bins=[0, 1, 3, 6, 12, 24, 60],
    labels=["<1m", "13m", "36m", "612m", "12y", "2y+"]
)
```

```python
retention_dist = (
    master
    .groupby(["cohort_month", "tenure_bucket"])
    .agg(customers=("account_id", "count"))
    .reset_index()
)
```

```python
retention_dist["cohort_size"] = retention_dist.groupby("cohort_month")["customers"].transform("sum")
retention_dist["retention_pct"] = retention_dist["customers"] / retention_dist["cohort_size"]
```

```python
retention_matrix = retention_dist.pivot(
    index="cohort_month",
    columns="tenure_bucket",
    values="retention_pct"
)
```
